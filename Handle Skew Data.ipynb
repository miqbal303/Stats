```python
# Load reuired library
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

# Right Skewed

```python
# Generate right-skewed data
R_skew_data = np.random.gamma(1, 2, 1000)

# Plot the distribution
plt.hist(R_skew_data, bins=50, color='green', alpha=0.5)
plt.title('Right-skewed Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Log transformation

```python
# histogram
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # specify horizontal and vertical spacing    
ax[0, 0].hist(R_skew_data, bins=50, color='blue', alpha=0.5)
ax[0, 0].set_title('Original Data Histogram')
ax[0, 0].set_xlabel('data')

# normal probability plot
stats.probplot(R_skew_data, dist='norm', plot=ax[0, 1])
ax[0, 1].set_title('Normal Probability Plot')
ax[0, 1].set_xlabel('Theoretical Quantiles')

# log transformation
log_variable = np.log(R_skew_data)
ax[1, 0].hist(log_variable, bins=50, color='green', alpha=0.5)
ax[1, 0].set_title('Log-Transformed Data Histogram')
ax[1, 0].set_xlabel('log_data')
stats.probplot(log_variable, dist='norm', plot=ax[1, 1])
ax[1, 1].set_title('Log Normal Probability Plot')
ax[1, 1].set_xlabel('Theoretical Quantiles')
```

### Square Root Transformation

```python
# histogram
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # specify horizontal and vertical spacing    
ax[0, 0].hist(R_skew_data, bins=50, color='blue', alpha=0.5)
ax[0, 0].set_title('Original Data Histogram')
ax[0, 0].set_xlabel('data')

# normal probability plot
stats.probplot(R_skew_data, dist='norm', plot=ax[0, 1])
ax[0, 1].set_title('Normal Probability Plot')
ax[0, 1].set_xlabel('Theoretical Quantiles')

# Squareroot transformation
sqrt_variable = np.sqrt(R_skew_data)  
ax[1, 0].hist(sqrt_variable, bins=50, color='green', alpha=0.5)
ax[1, 0].set_title('Square Root Transformation')
ax[1, 0].set_xlabel('sqrt_data')
stats.probplot(sqrt_variable, dist='norm', plot=ax[1, 1])
ax[1, 1].set_title('Square Root Normal Probability Plot')
ax[1, 1].set_xlabel('Theoretical Quantiles')
```

```python
# histogram
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # specify horizontal and vertical spacing    
ax[0, 0].hist(R_skew_data, bins=50, color='blue', alpha=0.5)
ax[0, 0].set_title('Original Data Histogram')
ax[0, 0].set_xlabel('data')

# normal probability plot
stats.probplot(R_skew_data, dist='norm', plot=ax[0, 1])
ax[0, 1].set_title('Normal Probability Plot')
ax[0, 1].set_xlabel('Theoretical Quantiles')

# exponential transformation
exp_variable = np.exp(R_skew_data) 
ax[1, 0].hist(exp_variable, bins=50, color='green', alpha=0.5)
ax[1, 0].set_title('Exponential Transformation')
ax[1, 0].set_xlabel('exp_data')
stats.probplot(exp_variable, dist='norm', plot=ax[1, 1])
ax[1, 1].set_title('Exponential Normal Probability Plot')
ax[1, 1].set_xlabel('Theoretical Quantiles')
```

###  Boxcox transformation

```python
# histogram
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # specify horizontal and vertical spacing    
ax[0, 0].hist(R_skew_data, bins=50, color='blue', alpha=0.5)
ax[0, 0].set_title('Original Data Histogram')
ax[0, 0].set_xlabel('data')

# normal probability plot
stats.probplot(R_skew_data, dist='norm', plot=ax[0, 1])
ax[0, 1].set_title('Normal Probability Plot')
ax[0, 1].set_xlabel('Theoretical Quantiles')

# boxcox transformation
boxcox_variable, _ = stats.boxcox(R_skew_data)
ax[1, 0].hist(boxcox_variable, bins=50, color='green', alpha=0.5)
ax[1, 0].set_title('Box-Cox Transformation')
ax[1, 0].set_xlabel('exp_data')
stats.probplot(boxcox_variable, dist='norm', plot=ax[1, 1])
ax[1, 1].set_title('Box-Cox Probability Plot')
ax[1, 1].set_xlabel('Theoretical Quantiles')
```
